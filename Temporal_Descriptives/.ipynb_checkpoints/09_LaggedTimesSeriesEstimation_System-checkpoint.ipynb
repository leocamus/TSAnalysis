### Summary of results: lagged_OLS_Summary

```python
lagged_OLS_Summary = pd.DataFrame()
```

```python
def summarizingResults(x, df, model_name):
    params = x.params.rename('params_' + model_name)
    t = x.tvalues.rename('t_' + model_name)
    p = x.pvalues.rename('p_' + model_name)

    df = pd.concat([df, params, t, p], axis=1)
    df.loc['R2_ADJ', 'params_' + model_name] = x.rsquared_adj
    df.loc['AIC', 'params_' + model_name] = x.aic
    df.loc['F', 'params_' + model_name] = x.fvalue
    df.loc['P_F', 'params_' + model_name] = x.f_pvalue
    df.loc['D_W', 'params_' + model_name] = sm.stats.stattools.durbin_watson(x.resid)
    
    return df
```

#### ESTIMATION. M11: just one lag

```python
complete_db['lag_ratio_tm'] = complete_db['ratio_tm'].shift(1)
```

```python
Y = complete_db.loc[:,'TOTAL_trx']
X = complete_db.loc[:,['SATURDAY',
                        'SUNDAY',
                        'lag_ratio_tm',
                        'kms_ofertados',
                        'WEEK_OF_YEAR_52',
                        'WEEK_OF_YEAR_53',
                        'WEEK_OF_YEAR_1',
                        'WEEK_OF_YEAR_2',
                        'WEEK_OF_YEAR_3',
                        'WEEK_OF_YEAR_4',
                        'WEEK_OF_YEAR_5',
                        'WEEK_OF_YEAR_6',
                        'WEEK_OF_YEAR_7',
                        'WEEK_OF_YEAR_8',
                        'WEEK_OF_YEAR_9',
                        'Julio',
                        't',
                        'Feriado_laboral',
                        'Feriado_no_laboral',
                        'FDS_Largo',
                        'Disturbios',
                        'Clima',
                        'Partido',
                        'Censo',
                        'visperas_laborales',
                        'N_ZPs']]

X = sm.add_constant(X)
m_11 = sm.OLS(Y, X, missing='drop')
results_11 = m_11.fit()
print(results_11.summary())
lagged_OLS_Summary = summarizingResults(results_11, lagged_OLS_Summary, 'm11')

ypred_m11 = results_11.predict(X)
complete_db.loc[:,'ypred_m11'] = ypred_m11
```

### Plotting

* Interactive, Daily

```python
def interactiveDailyPlotting(days_values, yact, ypred, model_name):
    trace1 = go.Scatter(x=days_values, y=yact, name='Actual', marker=dict(color='rgb(76, 181, 245)'),opacity=1)
    trace2 = go.Scatter(x=days_values, y=ypred, name='Predict', marker=dict(color='rgb(183, 184, 182)'),opacity=1)
    data = [trace1, trace2]
    layout = go.Layout(title='Transacciones por da (actual vs. predict.), ' + model_name,  yaxis=dict(title='Transacciones'))

    fig = go.Figure(data=data, layout=layout)
    plotly.offline.plot(fig, filename='act_vs_predict_day_' + model_name + '.html')
```

```python
days_values = complete_db.loc[:,'DATE']
yact = complete_db.loc[:,'TOTAL_trx']
```

* Interactive, Monthly

```python
def interactiveMonthlyPlotting(month_values, yact, ypred, model_name):
    trace1 = go.Scatter(x=month_values, y=yact, name='Actual', marker=dict(color='rgb(76, 181, 245)'),opacity=1)
    trace2 = go.Scatter(x=month_values, y=ypred, name='Predict', marker=dict(color='rgb(183, 184, 182)'),opacity=1)
    data = [trace1, trace2]
    layout = go.Layout(title='Transacciones por mes (actual vs. predict.), ' + model_name,  yaxis=dict(title='Transacciones'))

    fig = go.Figure(data=data, layout=layout)
    plotly.offline.plot(fig, filename='act_vs_predict_month_' + model_name + '.html')
```

```python
grouped_by_month = complete_db.groupby(pd.Grouper(key='DATE', freq='M'))
months_values = list(grouped_by_month.groups.keys())
yact = grouped_by_month['TOTAL_trx'].sum()
ypred_m11 = grouped_by_month['ypred_m11'].sum()
ypred_m12 = grouped_by_month['ypred_m12'].sum()
ypred_m13 = grouped_by_month['ypred_m13'].sum()
ypred_m14 = grouped_by_month['ypred_m14'].sum()
ypred_m15 = grouped_by_month['ypred_m15'].sum()
ypred_m16 = grouped_by_month['ypred_m16'].sum()
ypred_m17 = grouped_by_month['ypred_m17'].sum()
ypred_m18 = grouped_by_month['ypred_m18'].sum()
```

```python
interactiveMonthlyPlotting(months_values,yact,ypred_m11, 'M11')
interactiveMonthlyPlotting(months_values,yact,ypred_m12, 'M12')
interactiveMonthlyPlotting(months_values,yact,ypred_m13, 'M13')
interactiveMonthlyPlotting(months_values,yact,ypred_m14, 'M14')
interactiveMonthlyPlotting(months_values,yact,ypred_m15, 'M15')
interactiveMonthlyPlotting(months_values,yact,ypred_m16, 'M16')
interactiveMonthlyPlotting(months_values,yact,ypred_m17, 'M17')
interactiveMonthlyPlotting(months_values,yact,ypred_m18, 'M18')
```

* Static, Monthly

```python
colors = [(76, 181, 245),(183, 184, 182),(52, 103, 92),(179, 193, 0)]
for i in range(len(colors)):
    r, g, b = colors[i]  
    colors[i] = (r / 255., g / 255., b / 255.)
```

```python
def millions(x, pos):
    'The two args are the value and tick position'
    return '%1.1fM' % (x*1e-6)
```

```python
def staticMonthlyPlotting(month_values, yact, ypred, model_name):
    formatter = FuncFormatter(millions)
    fig, ax = plt.subplots()
    ax.plot(months_values, yact,'ro-' ,color = colors[0], label = 'Actual')
    ax.plot(months_values, ypred,'d-', color = colors[1], label = 'Predict')
    formatter = FuncFormatter(millions)
    ax.yaxis.set_major_formatter(formatter)
    ax.set_title('Transacciones - Actual vs. Predict., ' + model_name)
    ax.set_ylabel('Transacciones')
    ax.set_xlabel('Mes del ao')
    fig.autofmt_xdate()
    ax.legend(loc=3)

    plt.savefig('C:/Users/leoca_000/Desktop/Evasion/03_report/02_Figs/act_vs_predict_month_' + model_name + '.pdf')
```

```python
staticMonthlyPlotting(months_values,yact,ypred_m11, 'M11')
staticMonthlyPlotting(months_values,yact,ypred_m12, 'M12')
staticMonthlyPlotting(months_values,yact,ypred_m13, 'M13')
staticMonthlyPlotting(months_values,yact,ypred_m14, 'M14')
staticMonthlyPlotting(months_values,yact,ypred_m15, 'M15')
staticMonthlyPlotting(months_values,yact,ypred_m16, 'M16')
staticMonthlyPlotting(months_values,yact,ypred_m17, 'M17')
staticMonthlyPlotting(months_values,yact,ypred_m18, 'M18')
```
