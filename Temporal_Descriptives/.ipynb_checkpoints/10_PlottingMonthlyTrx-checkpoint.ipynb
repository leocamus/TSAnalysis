##### Creating function to analyze normal-stops by years

```python
def year_pn_analyses_trx(year,ana_turnstiles_df,mauricio_turnstiles_df):
    path = os.path.join(DTPM_TRXDir,'un_ppu_sersen_sumtrx_' + str(year) + '_by_date.csv')
    df = pd.read_csv(path, sep=";", header=None, encoding='latin-1', usecols=[0,1,2,3,4,5], parse_dates=[3])
    df.columns = ['UN','PPU','SER_SEN','DATE','SUM_TRX','COUNT']
    print('Number of observations before merge info. of turnstile is: ' + str(len(df.index)))
    
    df = df.merge(ana_turnstiles_df, left_on = 'PPU', right_on = 'sitio_subida', how='left', suffixes=('','_ana'))
    df = df.merge(mauricio_turnstiles_df, left_on = 'PPU', right_on = 'sitio_subida' , suffixes=('_ana', '_mauricio'), how='left')
    print('Number of observations after merge info. of turnstile is: ' + str(len(df.index)))
    
    torniquetes_mariposa_conditions = (df.loc[:,'fecha_instalacion_ana'].dt.date<df.loc[:,'DATE'].dt.date)
    df.loc[:,'min_fecha'] = pd.concat([df['fecha_instalacion_ana'], df['fecha_instalacion_mauricio']], axis=1).min(axis=1)
    no_torniquetes_conditions = (((df.loc[:,'fecha_instalacion_ana'].isnull()) & (df.loc[:,'fecha_instalacion_mauricio'].isnull())) | (df.loc[:,'DATE'].dt.date<=df['min_fecha'].dt.date))
    df.loc[:,'torniquete_mariposa'] = np.where(torniquetes_mariposa_conditions,1,0)
    df.loc[:,'no_torniquete'] = np.where(no_torniquetes_conditions,1,0)
    
    df.loc[:,'MONTH'] = df.loc[:,'DATE'].dt.month
    f = {'SUM_TRX':
         {'pn_SUM_TRX':['sum']},
         'COUNT':
         {'pn_SUM_EXP':['sum']}}
    grouped_df = df.groupby(['MONTH','torniquete_mariposa','no_torniquete']).agg(f)
    grouped_df.columns = grouped_df.columns.droplevel(1)
    grouped_df.reset_index(inplace=True,level=['torniquete_mariposa','no_torniquete'])
    months = grouped_df.groupby(['MONTH']).agg({'pn_SUM_EXP': 'sum'})
    grouped_df.loc[:,'ratio'] = grouped_df['pn_SUM_EXP'].div(months['pn_SUM_EXP'], level='MONTH') * 100
    
    return grouped_df
```

##### Creating function to analyze zp-stops years

```python
def year_zp_analyses_trx(year):
    path = os.path.join(DTPM_TRXDir,'trxzp_' + str(year))
    df = pd.read_csv(path, sep=";", header=None, encoding='latin-1', parse_dates=[2])
    df.columns = ['UN','RMZP','DATE','TIPODIA','MHORA','PERIODO','TRX_VALIDAS','TARJETAS_NO_VALIDAS','TRX_NO_VALIDAS']
    print('Number of observations is: ' + str(len(df.index)))
    
    df.loc[:,'MONTH'] = df.loc[:,'DATE'].dt.month
    f = {'TRX_VALIDAS':
         {'zp_SUM_TRX':['sum']},
        'TRX_NO_VALIDAS':
        {'zp_SUM_TRX_NO_VALIDAS':['sum']}}
    
    grouped_df = df.groupby(['MONTH']).agg(f)
    grouped_df.columns = grouped_df.columns.droplevel(1)

    return grouped_df
```

##### Colors and others settings for plotting

```python
colors = [(76, 181, 245),(183, 184, 182),(52, 103, 92),(179, 193, 0)]
for i in range(len(colors)):
    r, g, b = colors[i]  
    colors[i] = (r / 255., g / 255., b / 255.)
```

```python
def millions(x, pos):
    'The two args are the value and tick position'
    return '%1.1fM' % (x*1e-6)
```

```python
from matplotlib.ticker import FuncFormatter
formatter = FuncFormatter(millions)
```

##### Analyzing 2015

```python
tic = time.clock()

pn_grouped_2015_df = year_pn_analyses_trx(2015,ana_turnstiles_df,mauricio_turnstiles_df)
zp_grouped_2015_df = year_zp_analyses_trx(2015)

toc = time.clock()
print('Time of operations is: ' + str(toc-tic))
```

```python
no_turnstile_2015 = pn_grouped_2015_df.loc[(pn_grouped_2015_df['torniquete_mariposa']==0)
                                           &(pn_grouped_2015_df['no_torniquete']==1),:]

three_turnstile_2015 = pn_grouped_2015_df.loc[(pn_grouped_2015_df['torniquete_mariposa']==0)
                                              &(pn_grouped_2015_df['no_torniquete']==0),:]

butterfly_turnstile_2015 = pn_grouped_2015_df.loc[(pn_grouped_2015_df['torniquete_mariposa']==1)
                                                  &(pn_grouped_2015_df['no_torniquete']==0),:]
```

```python
butterfly_turnstile_2015
```

```python
for i in range(1,5):
    butterfly_turnstile_2015.loc[i,['torniquete_mariposa','no_torniquete','pn_SUM_TRX','pn_SUM_EXP','ratio']] = ['1','0','0','0','0']
    
butterfly_turnstile_2015.sort_index(inplace=True)
```

```python
butterfly_turnstile_2015
```

```python
print(no_turnstile_2015.index.is_monotonic)
print(three_turnstile_2015.index.is_monotonic)
print(butterfly_turnstile_2015.index.is_monotonic)
```

```python
months_values = no_turnstile_2015.index.values
no_turnstile_2015_values = list(map(int, no_turnstile_2015['pn_SUM_TRX'].values))
three_turnstile_2015_values = list(map(int,three_turnstile_2015['pn_SUM_TRX'].values))
butterfly_turnstile_2015_values = list(map(int,butterfly_turnstile_2015['pn_SUM_TRX'].values))
zp_2015_values = list(map(int,zp_grouped_2015_df['zp_SUM_TRX'].values))
ratio_butterfly_turnstile_2015_values = list(map(int,butterfly_turnstile_2015['ratio'].values))
```

```python
labels = ["zp", "S/T", "C/T3", "C/TM"]
fig, ax = plt.subplots()
ax.stackplot(months_values, zp_2015_values, no_turnstile_2015_values , three_turnstile_2015_values , butterfly_turnstile_2015_values , labels = labels , colors=colors)
ax.legend(loc=2)
ax.set_title('Transacciones por mes, 2015')
ax.set_ylabel('Transacciones')
ax.yaxis.set_major_formatter(formatter)
ax.set_xlabel('Mes del a√±o')

ax_1 = ax.twinx()
ax_1.plot(months_values, ratio_butterfly_turnstile_2015_values, 'ro-', color = 'black',  label = '% Exp.')
ax_1.set_ylabel('% Expediciones con torniquete mariposa')
ax_1.legend(loc=4)
ax_1.set_ylim(0, 100)

plt.xticks(np.arange(min(months_values), max(months_values)+1, 1.0))
#plt.savefig('C:/Users/leoca_000/Desktop/Evasion/03_report/02_Figs/5_trxOverTimeByMonth_2015.pdf')
plt.show()
```
