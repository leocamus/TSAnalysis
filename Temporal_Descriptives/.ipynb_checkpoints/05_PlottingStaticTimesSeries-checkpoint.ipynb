```python
%matplotlib inline
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.ticker import FuncFormatter
import numpy as np
import datetime as dt
import time
```

```python
from Utils import TransantiagoConstants
DTPM_TRXDir = TransantiagoConstants.DTPM_TRXDir
```

* Defining global things

```python
colors = [(76, 181, 245),(183, 184, 182),(52, 103, 92),(179, 193, 0)]
for i in range(len(colors)):
    r, g, b = colors[i]  
    colors[i] = (r / 255., g / 255., b / 255.)
```

```python
def millions(x, pos):
    'The two args are the value and tick position'
    return '%1.1fM' % (x*1e-6)
```

* Defining functions to plot

```python
def monthlyPlotter(save_figure):
    
    monthly_input_path = os.path.join(DTPM_TRXDir,'1_MONTHLY/monthly_summary.csv')
    monthly_trx = pd.read_csv(monthly_input_path,sep=';',encoding='latin-1', index_col=0)
    
    months_values = pd.date_range(start='2015-01',end='2018-01',freq='M').values
    no_turnstile_values = list(map(int, monthly_trx['pn_SUM_TRX_no_t'].values))
    three_turnstile_values = list(map(int,monthly_trx['pn_SUM_TRX_3t'].values))
    butterfly_turnstile_values = list(map(int,monthly_trx['pn_SUM_TRX_tm'].values))
    zp_values = list(map(int,monthly_trx['zp_SUM_TRX'].values))
    ratio_butterfly_turnstile_values = list(map(int,monthly_trx['ratio_tm'].values))
    
    formatter = FuncFormatter(millions)
    
    labels = ["ZP", "S/T", "C/T3", "C/TM"]
    fig, ax = plt.subplots()
    ax.stackplot(months_values, zp_values, no_turnstile_values , three_turnstile_values , butterfly_turnstile_values , labels = labels , colors=colors)
    ax.legend(loc=2)
    ax.set_title('Transacciones por mes, 2015 a 2017')
    ax.set_ylabel('Transacciones')
    ax.yaxis.set_major_formatter(formatter)
    ax.set_xlabel('Mes del a単o')
    fig.autofmt_xdate()
    
    ax_1 = ax.twinx()
    ax_1.plot(months_values, ratio_butterfly_turnstile_values, 'ro-', color = 'black',  label = '% Exp.')
    ax_1.set_ylabel('% Expediciones con torniquete mariposa')
    ax_1.legend(loc=4)
    ax_1.set_ylim(0, 100)

    if save_figure:
        plt.savefig('C:/Users/leoca_000/Desktop/Evasion/03_report/02_Figs/5_trxOverTimeByMonth_all.pdf')
        
    plt.show()
```
