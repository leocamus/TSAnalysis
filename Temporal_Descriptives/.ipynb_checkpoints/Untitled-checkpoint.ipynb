```python
import statsmodels.api as sm
```

```python
from Utils import TransantiagoConstants
```

```python
fourLevelsUp = TransantiagoConstants.fourLevelsUp
```

```python
travel_time_path = os.path.join(fourLevelsUp, '03_datos/12_TRAVEL_TIME/base_modelo.csv')
travel_time = pd.read_csv(travel_time_path, sep=';', encoding='latin-1')
```

```python
Y = travel_time.loc[:,'TEXP']
X = travel_time.loc[:,['largo_ruta','NI','Y','DET']]
X = sm.add_constant(X)
m1 = sm.OLS(Y, X)
results_m1 = m1.fit()
print(results_m1.summary())
```

```python
travel_time.loc[:,'Y_2'] = 2*travel_time.loc[:,'Y']
```

```python
Y = travel_time.loc[:,'TEXP']
X = travel_time.loc[:,['largo_ruta','NI','Y_2','DET']]
X = sm.add_constant(X)
m1 = sm.OLS(Y, X)
results_m1 = m1.fit()
print(results_m1.summary())
```

```python
anexo_4_path = os.path.join(fourLevelsUp, '03_datos/04_DTPM/02_Anexos_4/2017-11-24_consolidado_anexo4_(Circunvalaci√≥n).xlsx')
anexo_4 = pd.read_excel(anexo_4_path)
```

```python
anexo_4 = anexo_4.fillna('')
```

```python
anexo_4.loc[:,'CODE_TS'] = anexo_4.loc[:,'Servicio Usuario'].map(str) + anexo_4.loc[:,'Sentido Servicio'] + anexo_4.loc[:,'Varian-te']
```

```python
grouped_anexo_4 = anexo_4.groupby(['CODE_TS']).size().to_frame().reset_index()
```

```python
travel_time = travel_time.merge(grouped_anexo_4, left_on=['COD_USU_SENTIDO'], right_on = ['CODE_TS'], how='left')
```

```python
travel_time.loc[:,'DIF'] = travel_time.loc[:,'DET'] - travel_time.loc[:,0]
```

```python
travel_time = travel_time.sort_values(by=['DIF'], ascending=False)
```

```python
travel_time.head()
```

```python
travel_time_2 = travel_time.loc[travel_time['DIF']<=0,:]
```

```python
Y = travel_time_2.loc[:,'TEXP']
X = travel_time_2.loc[:,['largo_ruta','NI','Y','DET']]
X = sm.add_constant(X)
m1 = sm.OLS(Y, X)
results_m1 = m1.fit()
print(results_m1.summary())
```
