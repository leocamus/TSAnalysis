```python
etapas_df = RunDailyEtapasBuilder.loadSimplifiedEtapas()
perfiles_df = RunDailyEtapasBuilder.loadSimplifiedPerfiles()
```

```python
etapas_df.loc[:,'t_subida'] = pd.to_datetime(etapas_df.t_subida)
perfiles_df.loc[:,'Hini'] = pd.to_datetime(perfiles_df.Hini)
perfiles_df.loc[:,'Hfin'] = pd.to_datetime(perfiles_df.Hfin)
```

```python
etapas_df.head()
```

```python
perfiles_df.head()
```

```python
columns_names = ['ServicioSentido','Patente','idExpedicion']
simplified_perfiles_df = perfiles_df.drop_duplicates(columns_names)
```

* Below line takes too long. Do not run it yet

```python
etapas_df['idExpedicion']=''

for etapas_index, etapas_row in etapas_df.iterrows():
    patente = etapas_row['sitio_subida']
    servicio = etapas_row['servicio_subida']
    t_subida = etapas_row['t_subida']    
    util_perfil_rows = simplified_perfiles_df[(simplified_perfiles_df['Patente']==patente)&
                                             (simplified_perfiles_df['ServicioSentido']==servicio)]
#    if util_perfil_rows.empty:
#        print('something is wrong')
        
    for perfiles_index, perfiles_row in util_perfil_rows.iterrows():
        Hini = perfiles_row['Hini']
        Hfin = perfiles_row['Hfin']
        if(Hini<=t_subida<=Hfin):
            etapas_df.loc[etapas_index,'idExpedicion'] = perfiles_row['idExpedicion']
            break
```

```python
simplified_perfiles_df = simplified_perfiles_df.sort_values(by=['Patente','ServicioSentido','idExpedicion'], ascending=[True,True, True])
```

```python
simplified_perfiles_df = simplified_perfiles_df.reset_index()
```

```python
simplified_perfiles_df.head(50)
```

```python
simplified_perfiles_df['misma_patente'] = (simplified_perfiles_df['Patente']==simplified_perfiles_df['Patente'].shift()).fillna(False)
simplified_perfiles_df['mismo_servicio'] = (simplified_perfiles_df['ServicioSentido']==simplified_perfiles_df['ServicioSentido'].shift()).fillna(False)
```

```python
simplified_perfiles_df.head()
```

```python
simplified_perfiles_df.loc[(simplified_perfiles_df.misma_patente == True) &
                           (simplified_perfiles_df.mismo_servicio == True), 'diferencia_tiempo_exp'] = (simplified_perfiles_df['Hini']-simplified_perfiles_df['Hfin'].shift())
```

```python
simplified_perfiles_df.head()
```

```python
not_null_simplified_perfiles = simplified_perfiles_df[simplified_perfiles_df['diferencia_tiempo_exp'].notnull()]
```
